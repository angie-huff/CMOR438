# Decision and Regression Trees

## Introduction

This notebook explores the use of both **decision trees** and **regression trees** to model the Openness to Experience dimension of personality, one of the Big Five personality traits. These two models belong to the family of tree-based supervised learning algorithms, known for their interpretability and ability to model non-linear patterns in the data.

A **decision tree classifier** is designed for categorical prediction. In this context, the model classifies individuals as either high or low in openness based on their responses to ten survey items (OPN1 through OPN10). The tree is constructed by recursively partitioning the dataset based on feature thresholds that maximize the purity of the resulting subsets. At each internal node, the algorithm selects a feature and a split point that best separate the data into more homogeneous groups, using criteria such as Gini impurity. The final leaf nodes represent the predicted class labels.

A **regression tree**, on the other hand, is used to predict continuous values. Instead of classifying individuals into high or low openness, the regression tree estimates their exact average openness score. Like the classifier, it splits the data into increasingly homogeneous subsets, but instead of maximizing class purity, it minimizes the variance of the target variable within each split. The predicted value at each leaf is the average of the target values for samples that fall into that region.

Using the same dataset for both models allows for a direct comparison between classification and regression. The decision tree section evaluates accuracy and interpretability using confusion matrices and classification metrics. The regression tree section assesses performance using mean squared error and predicted vs. actual plots.

This notebook demonstrates the versatility of tree-based models and how they can be applied to both categorical and continuous prediction problems in psychological data analysis.


### Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

### Loading and Preprocessing the Data

We use the same openness dataset and preprocessing steps applied in the KNN notebook to ensure comparability across models. 

The dataset contains 10 survey items related to Openness to Experience (OPN1–OPN10). We calculate each individual's mean openness score and assign a binary label: 1 for high openness (above the median), and 0 for low openness (below or equal to the median). Features are standardized to improve model interpretability and ensure consistent treatment across classifiers.


```python
# Load dataset and sample 2,000 rows
df = pd.read_csv("Personality.csv", sep='\t')
df = df.sample(n=2000, random_state=42)

# Use only the 10 Openness items (OPN1 to OPN10)
openness_cols = [f"OPN{i}" for i in range(1, 11)]
data = df[openness_cols].dropna()

# Create binary labels: 1 = high openness, 0 = low openness
openness_score = data.mean(axis=1)
median_score = openness_score.median()
labels = (openness_score > median_score).astype(int)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, labels, test_size=0.2, random_state=42
)
```

### Training and Visualizing the Decision Tree

We train a `DecisionTreeClassifier` using the training data and set the maximum depth to 4. This limits the complexity of the model, helping to prevent overfitting and improving interpretability.

The tree diagram below shows how the model splits the dataset based on different openness features. Each internal node represents a decision rule (e.g., OPN8 ≤ 0.266), and each leaf node represents a predicted class (Low or High openness).

The `gini` value at each node indicates the impurity of the split, where 0.0 represents perfect classification. The `samples` value indicates how many data points reach that node, and the `value` shows the class distribution at that node.

From the plot, we can observe:
- OPN8 and OPN1 are used near the root, indicating they are the most informative features in the tree.
- The tree consistently splits to isolate samples with low impurity (e.g., `gini = 0.0` or near-zero leaves), showing the model is effectively separating high and low openness groups.


```python
# Train a decision tree classifier with limited depth
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train, y_train)

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(clf, filled=True, feature_names=openness_cols, class_names=["Low", "High"], rounded=True)
plt.title("Decision Tree (max_depth=4)")
plt.show()
```

### Model Evaluation with Confusion Matrix and Classification Report

We evaluate the decision tree classifier using a confusion matrix and classification report on the test set.

The model achieves an overall accuracy of 84%, with reasonably balanced performance across both classes. Precision and recall scores indicate that the classifier is slightly better at identifying low-openness individuals, but still performs well on high-openness cases.

- **Class 0 (Low Openness):**
  - Precision: 0.88
  - Recall: 0.85
  - F1-score: 0.86
- **Class 1 (High Openness):**
  - Precision: 0.79
  - Recall: 0.82
  - F1-score: 0.80

These results show that the decision tree captures the underlying patterns in the data well, with slightly better performance on the majority class (Low).



```python
# Predict on test set
y_pred = clf.predict(X_test)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix plot
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Low', 'High'], yticklabels=['Low', 'High'])
plt.title("Confusion Matrix (Decision Tree)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

### Preparing Data for Regression

To build a regression tree, we replace the binary classification labels with each individual's actual mean openness score. This allows the model to predict a continuous value rather than a class. The feature matrix remains the same, and we continue using standardized input features.

```python
# Use the same feature matrix
X = X_scaled

# Target is now the continuous openness score
y_continuous = openness_score.values

# Train-test split for regression task
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X, y_continuous, test_size=0.2, random_state=42
)
```

### Training and Visualizing the Regression Tree

We train a `DecisionTreeRegressor` to predict the average openness score for each individual. The tree structure is similar to that of a classifier, but the predicted values at each leaf are continuous.

```python
# Train a regression tree
reg_tree = DecisionTreeRegressor(max_depth=3, random_state=42)
reg_tree.fit(X_train_r, y_train_r)

# Visualize the tree
plt.figure(figsize=(20, 10))
plot_tree(reg_tree, filled=True, feature_names=openness_cols, rounded=True)
plt.title("Regression Tree (max_depth=3)")
plt.show()
```

The regression tree shown above was trained to predict the average openness score for each individual based on their responses to ten survey items (OPN1–OPN10). The tree was constrained to a maximum depth of 3 to improve interpretability.

Each internal node contains:
- A decision rule on one of the openness features
- The number of samples in that node
- The average openness score for those samples (`value`)
- The mean squared error (`squared_error`) within that subset

Each leaf node represents a predicted openness score for individuals that fall into that region of the feature space. For example, if a person’s OPN8 score is greater than 0.266 and their OPN5 score is less than or equal to −1.283, the model would predict their openness to be approximately 3.405.

This tree demonstrates how the regression model partitions the feature space into regions where the openness scores are relatively homogeneous. Smaller `squared_error` values in the leaves indicate that the model is capturing well-separated segments of the population with similar openness levels.


### Evaluating the Regression Tree

To evaluate the performance of the regression tree, we compute the **Mean Squared Error (MSE)** between the predicted and actual openness scores on the test set. MSE quantifies the average squared difference between predicted and true values — lower values indicate better predictive accuracy.

We also plot predicted values against actual values to visually assess how well the model captures the continuous range of openness scores. Ideally, points should lie close to the identity line $y = x$, indicating near-perfect predictions.

```python
# Predict on the test set
y_pred_r = reg_tree.predict(X_test_r)

# Compute Mean Squared Error
mse = mean_squared_error(y_test_r, y_pred_r)
print(f"Mean Squared Error (MSE): {mse:.4f}")

# Scatterplot of predicted vs actual
plt.figure(figsize=(8, 6))
plt.scatter(y_test_r, y_pred_r, alpha=0.6, edgecolors='k')
plt.plot([y_test_r.min(), y_test_r.max()], [y_test_r.min(), y_test_r.max()], 'r--')
plt.xlabel("Actual Openness Score")
plt.ylabel("Predicted Openness Score")
plt.title("Predicted vs. Actual Openness Scores")
plt.grid(True)
plt.tight_layout()
plt.show()
```

The regression tree achieved an MSE of 0.0937 on the test set. This suggests that, on average, the model's predictions are very close to the actual openness scores.

The scatterplot of predicted vs. actual scores reveals that the model captures the general structure of the data but tends to cluster predictions near the overall mean openness value. While the red dashed line represents perfect prediction ($y = x$), many points fall slightly below it — suggesting some underestimation for individuals with higher openness.

This outcome is expected with a tree of limited depth (max_depth = 3), as regression trees create piecewise constant approximations. Increasing the tree depth would likely improve accuracy but reduce interpretability and increase the risk of overfitting.

Overall, the regression tree offers a solid baseline for continuous prediction, capturing central tendencies effectively while maintaining a low error rate.

## Conclusion

In this notebook, we explored both **decision trees** and **regression trees** as interpretable, tree-based models for analyzing Openness to Experience using survey data.

The **decision tree classifier** was used to predict whether individuals scored above or below the median openness level. With a maximum depth of 4, the model achieved an accuracy of **84%**, showing balanced precision and recall across both classes. The confusion matrix and tree visualization revealed meaningful feature splits and confirmed that the model could distinguish high- and low-openness individuals with relatively few misclassifications.

The **regression tree** was used to predict each individual’s exact mean openness score as a continuous value. With a maximum depth of 3, the model produced a **mean squared error (MSE) of 0.0937**, and the predicted vs. actual scatterplot showed that the tree approximated openness scores well, though predictions tended to cluster around the mean.

Together, these models demonstrate the flexibility and interpretability of tree-based approaches. Decision trees provided discrete categorization for classification tasks, while regression trees offered a simple but effective piecewise approximation for continuous prediction. Both models serve as transparent baselines that balance predictive accuracy with intuitive structure, making them especially suitable for domains like psychology where model explainability is key.
