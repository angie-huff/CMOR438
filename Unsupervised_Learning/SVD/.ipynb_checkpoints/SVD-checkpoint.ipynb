# Singular Value Decomposition (SVD)

In this notebook, we apply **Singular Value Decomposition (SVD)** to explore patterns in high-dimensional mental health survey data. Like PCA, SVD is a powerful technique for **dimensionality reduction**, commonly used in unsupervised learning to reveal underlying structure in large datasets.

### What is SVD?

Singular Value Decomposition (SVD) factorizes any real-valued matrix $X$ into three components:

$$
X = U \Sigma V^T
$$

- $U$: Left singular vectors (rows = observations)
- $\Sigma$: Diagonal matrix of singular values (importance of each component)
- $V^T$: Right singular vectors (columns = original features)

In the context of data science, we use a variant called **Truncated SVD**, which:
- Approximates the original matrix using only the top $k$ components
- Is especially useful for **sparse or non-centered** data (unlike PCA which requires centering)
- Produces low-dimensional representations while preserving the maximum variance possible

Here, we use SVD to:
- Reduce our 26-feature mental health dataset to 2 dimensions
- Visualize variation across students
- Compare with PCA and assess the structure of mental health symptom profiles

### Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load and Preprocess Data
We use the same dataset from previous notebooks: the **"MHP (Anxiety, Stress, Depression) Dataset of University Students"**, which contains self-reported mental health survey responses from **2,028 university students** across Bangladesh.

**About the Dataset:**

- **Source**: [Figshare – MHP Dataset](https://doi.org/10.6084/m9.figshare.25771164.v1)
- **Collection Year**: 2024
- **Population**: Undergraduate and graduate students, all academic years and majors
- **Instrument**: A structured questionnaire adapted from validated psychological tools like PHQ-9 and GAD-7, along with original academic stress items
- **Response Format**: 5-point Likert scale for each item (e.g., “0 - Not at all” to “4 - Very Often”)

**Features Used:**

From the full dataset, we select **26 Likert-scale items** specifically related to:

- **Academic stress** (e.g., coping with assignments and performance pressure)
- **Anxiety symptoms** (e.g., restlessness, worry, nervousness)
- **Depression symptoms** (e.g., fatigue, suicidal ideation, loss of interest)

Each question measures symptom **frequency over the course of a semester** and is mapped to numerical values (0–4) to enable matrix computations.

**Preprocessing Steps:**

1. **Likert Mapping**: We convert categorical Likert responses into numeric scale values using a consistent mapping.
2. **Missing Value Handling**: Responses marked "Prefer not to say" or similar are treated as missing and removed.
3. **Standardization**: All features are scaled using `StandardScaler` to ensure zero mean and unit variance — a necessary condition for SVD-based dimensionality reduction.

This standardized data matrix is now ready for decomposition using **Truncated SVD**, which will allow us to uncover underlying structures in symptom patterns across students.

```python
df = pd.read_csv("MentalHealth.csv")

likert_columns = [
    '1. In a semester, how often have you felt upset due to something that happened in your academic affairs? ',
    '2. In a semester, how often you felt as if you were unable to control important things in your academic affairs?',
    '3. In a semester, how often you felt nervous and stressed because of academic pressure? ',
    '4. In a semester, how often you felt as if you could not cope with all the mandatory academic activities? (e.g, assignments, quiz, exams) ',
    '5. In a semester, how often you felt confident about your ability to handle your academic / university problems?',
    '6. In a semester, how often you felt as if things in your academic life is going on your way? ',
    '7. In a semester, how often are you able to control irritations in your academic / university affairs? ',
    '8. In a semester, how often you felt as if your academic performance was on top?',
    '9. In a semester, how often you got angered due to bad performance or low grades that is beyond your control? ',
    '10. In a semester, how often you felt as if academic difficulties are piling up so high that you could not overcome them? ',
    '1. In a semester, how often you felt nervous, anxious or on edge due to academic pressure? ',
    '2. In a semester, how often have you been unable to stop worrying about your academic affairs? ',
    '3. In a semester, how often have you had trouble relaxing due to academic pressure? ',
    '4. In a semester, how often have you been easily annoyed or irritated because of academic pressure?',
    '5. In a semester, how often have you worried too much about academic affairs? ',
    '6. In a semester, how often have you been so restless due to academic pressure that it is hard to sit still?',
    '7. In a semester, how often have you felt afraid, as if something awful might happen?',
    '1. In a semester, how often have you had little interest or pleasure in doing things?',
    '2. In a semester, how often have you been feeling down, depressed or hopeless?',
    '3. In a semester, how often have you had trouble falling or staying asleep, or sleeping too much? ',
    '4. In a semester, how often have you been feeling tired or having little energy? ',
    '5. In a semester, how often have you had poor appetite or overeating? ',
    '6. In a semester, how often have you been feeling bad about yourself - or that you are a failure or have let yourself or your family down? ',
    '7. In a semester, how often have you been having trouble concentrating on things, such as reading the books or watching television? ',
    "8. In a semester, how often have you moved or spoke too slowly for other people to notice? Or you've been moving a lot more than usual because you've been restless? ",
    '9. In a semester, how often have you had thoughts that you would be better off dead, or of hurting yourself? '
]

likert_mapping = {
    "0 - Not at all": 0, "0 - Never": 0,
    "1 - Several days": 1, "1 - Several days (less than 15 days)": 1, "1 - Almost Never": 1,
    "2 - Sometimes": 2, "2 - More than half the days": 2, "2 - More than half the semester": 2,
    "3 - Fairly Often": 3, "3 - Nearly every day": 3,
    "4 - Very Often": 4, "4 - Prefer not to say": np.nan
}

# Apply mapping and clean
df_numeric = df[likert_columns].replace(likert_mapping).infer_objects()
df_clean = df_numeric.dropna()

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean)
```

### Interpreting the SVD Explained Variance

The plot below shows how much variance is cumulatively explained by the top singular vectors in the data.

- The **first few components** explain a significant portion of the variance
- This justifies reducing the original 26-dimensional space to a smaller number of components (e.g., 2 or 3) while preserving the majority of information
- In our case, the **first 10 components** explain over 80% of the total variance

This validates the use of **2D SVD projections** for visualization and downstream analysis of student mental health profiles.

```python
# Apply Truncated SVD
svd = TruncatedSVD(n_components=26, random_state=42)
X_svd = svd.fit_transform(X_scaled)

# Calculate explained variance ratio
explained_variance = svd.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# Plot cumulative explained variance
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.title("SVD: Cumulative Explained Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance Explained")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualizing Students in SVD-Reduced Space

After applying Truncated Singular Value Decomposition (SVD), we now visualize the distribution of students across the first two SVD components.

Instead of a traditional scatter plot, we use a density map to capture how student responses to mental health symptoms are distributed in this reduced-dimensional space. This approach helps reveal subtle patterns and gradients in student psychological profiles that may not be evident from raw scatter points alone.

```python
# Re-run SVD with only 2 components for visualization
svd_2d = TruncatedSVD(n_components=2, random_state=42)
X_svd_2d = svd_2d.fit_transform(X_scaled)

# Build DataFrame for plotting
svd_df = pd.DataFrame(X_svd_2d, columns=["SVD1", "SVD2"])

plt.figure(figsize=(8, 6))
sns.kdeplot(
    x=svd_df["SVD1"], 
    y=svd_df["SVD2"], 
    fill=True, cmap="Blues", thresh=0.05, levels=100
)
plt.title("Density Map of Students in SVD-Reduced Space")
plt.xlabel("SVD Component 1")
plt.ylabel("SVD Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Conclusion

In this notebook, we used Truncated Singular Value Decomposition (SVD) to reduce the dimensionality of student mental health data and uncover its underlying structure.

### Key Findings:

- The **first two SVD components** captured a meaningful portion of the dataset's total variance, validating their use for 2D visualization.
- A **density plot** of the SVD projection showed a clear concentration of students near the center, with a gradual tapering toward the periphery. This supports the interpretation that **mental health symptoms vary continuously**, rather than existing in distinct, isolated categories.
- Outliers and edge-density regions may represent students with **unusual or severe symptom patterns**, warranting further investigation or targeted support.

### Broader Implications:

- SVD provides an interpretable low-dimensional embedding of complex psychological survey data, helping visualize the **spectrum of academic stress, anxiety, and depression**.
- These insights could be used by mental health practitioners or university administrators to better understand the **distribution of risk** within student populations.
- Finally, the results reaffirm that data-driven methods like SVD can play a powerful role in **psychological profiling, resource allocation, and early intervention strategies**.

In summary, SVD enabled us to reduce dimensionality without losing key structural information, and to visualize student mental health variation in a way that is both compact and socially meaningful.
