# $k$-Means Clustering

In this notebook, we use $k$-Means clustering to identify latent mental health profiles among university students using data from the *Mental Health Assessment of University Students* dataset. This dataset includes responses from over 2,000 students across various universities, academic years, and disciplines. Participants answered a series of questions related to stress, anxiety, and depression using a standardized Likert scale.

Our goal is to segment students into meaningful psychological profiles based on their responses to survey items about:
- Academic stress (e.g., workload, coping, confidence)
- Anxiety symptoms (e.g., nervousness, restlessness, fear)
- Depression symptoms (e.g., hopelessness, fatigue, suicidal ideation)

These groupings can help surface hidden patterns that could inform student wellness interventions or further hypothesis-driven research.

$k$-Means is a partition-based clustering algorithm that groups data into $K$ non-overlapping clusters based on feature similarity. Each cluster is defined by its centroid, and the algorithm aims to minimize the within-cluster sum of squares (WCSS) — i.e., how far each point is from the center of its assigned cluster.

Formally, given a dataset $X = \{x_1, x_2, ..., x_n\}$, the algorithm solves:

$$
\underset{C}{\text{argmin}} \sum_{k=1}^{K} \sum_{x \in C_k} \|x - \mu_k\|^2
$$

Where:
- $C_k$ is the set of points in cluster $k$
- $\mu_k$ is the centroid of cluster $k$
- $\|x - \mu_k\|^2$ is the squared Euclidean distance

The algorithm proceeds iteratively:

1. **Initialize** $k$ centroids (random or $k$-Means++)
2. **Assign** each data point to its nearest centroid
3. **Update** centroids as the mean of all assigned points:
   $$
   \mu_k = \frac{1}{|C_k|} \sum_{x \in C_k} x
   $$
4. Repeat steps 2–3 until convergence (no changes in assignments or a max number of iterations)

### Dataset and Features

The dataset includes 33 columns: demographic attributes and 25+ self-reported psychological items. Each mental health indicator is phrased as a question like:

> *“In a semester, how often have you had trouble falling or staying asleep, or sleeping too much?”*

Responses are encoded using a 5-point Likert scale, such as:
- `0 - Not at all`
- `1 - Several days`
- `2 - More than half the days`
- `3 - Nearly every day`
- `4 - Prefer not to say` (to be excluded)

For this notebook, we:
- Focus on a subset of numeric Likert-scale mental health questions
- Map responses to integer values
- Standardize features for Euclidean distance calculations
- Cluster the students using $k$-Means, an unsupervised algorithm

### Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

### Load and Preprocess Data
We now select a subset of Likert-scale questions that assess psychological well-being across three major domains:
- **Academic stress and coping**
- **Anxiety symptoms**
- **Depression symptoms**

These items use a 5-point Likert scale (e.g., “0 - Not at all” to “3 - Nearly every day”) that we convert to numerical scores for clustering.

We drop demographic features (e.g., age, university, CGPA) and focus only on the quantitative mental health responses. We also remove responses labeled “4 - Prefer not to say” to preserve numeric consistency.

```python
# Load dataset
df = pd.read_csv("MentalHealth.csv")

# Define relevant Likert-based columns for clustering
likert_columns = df.columns[7:]

# Define mapping from Likert text to numeric values
likert_mapping = {
    "0 - Not at all": 0,
    "0 - Never": 0,
    "1 - Several days": 1,
    "1 - Several days (less than 15 days)": 1,
    "1 - Almost Never": 1,
    "2 - Sometimes": 2,
    "2 - More than half the days": 2,
    "2 - More than half the semester": 2,
    "3 - Fairly Often": 3,
    "3 - Nearly every day": 3,
    "4 - Very Often": 4,
    "4 - Prefer not to say": np.nan
}

# Apply mapping
df_numeric = df[likert_columns].replace(likert_mapping)

# Drop rows with missing values from "Prefer not to say"
df_numeric_clean = df_numeric.dropna()
```

### Feature Scaling and Elbow Method

$k$-Means uses Euclidean distance to assign points to the nearest cluster. Therefore, it is crucial to standardize the data so that features with larger magnitudes do not dominate the distance calculation.

We use `StandardScaler` from `sklearn.preprocessing` to scale each feature to have:
- Mean = 0
- Standard deviation = 1

After scaling, we apply the **Elbow Method** to determine a good number of clusters $k$. This method fits $k$-Means for a range of $k$ values and plots the **inertia** (i.e., within-cluster sum of squared distances) for each.

The **inertia** is computed as:

$$
\text{Inertia} = \sum_{i=1}^{k} \sum_{x_j \in C_i} \|x_j - \mu_i\|^2
$$

Where:
- $C_i$ is the set of points in cluster $i$
- $\mu_i$ is the centroid of cluster $i$
- $\|x_j - \mu_i\|^2$ is the squared Euclidean distance

We select the $k$ at which inertia starts to level off — the "elbow" point — as a balance between compact clusters and model simplicity.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numeric_clean)

# Try different values of k
inertias = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

### PCA Visualization of Clusters

To visualize the clusters, we reduced the 26-dimensional data to 2 principal components using PCA.

Each point represents a student, color-coded by their K-Means cluster assignment:

- **Cluster 0 (leftmost)**: High distress students — appear tightly grouped but distinct
- **Cluster 1 (middle)**: Moderate symptom group — more spread out
- **Cluster 2 (rightmost)**: Low distress — also distinct, occupying their own space

This confirms that K-Means successfully separated students based on psychological similarity in symptom patterns.

PCA is a dimensionality reduction technique that transforms correlated input features into a new set of uncorrelated axes, called principal components. These components are linear combinations of the original features and are ranked by how much variance they explain.

- **PC1** captures the direction of maximum variance in the data.
- **PC2** captures the next orthogonal direction of maximum variance.
- The result is a compressed view of the data that preserves as much structure as possible.

Mathematically, PCA computes the **eigenvectors** and **eigenvalues** of the covariance matrix of the data:

- Eigenvectors define the new axes (principal components).
- Eigenvalues measure the variance captured by each axis.

We apply PCA to our standardized dataset and project each data point into a 2D space using the first two principal components.

```python
kmeans_final = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_final.fit(X_scaled)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df["Cluster"] = kmeans_final.labels_

# Visualize clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="Cluster", palette="Set2", alpha=0.7, s=60)
plt.title("K-Means Clusters Visualized Using PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Cluster")
plt.grid(True)
plt.tight_layout()
plt.show()
```
