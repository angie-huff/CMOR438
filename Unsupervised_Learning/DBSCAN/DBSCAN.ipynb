# DBSCAN Clustering

In this section, we apply **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise), a clustering algorithm that groups data based on density rather than distance to centroids.

Unlike K-Means, DBSCAN:
- Does not require specifying the number of clusters in advance
- Can find clusters of arbitrary shape
- Identifies noise points (outliers that don’t belong to any cluster)

DBSCAN defines clusters using two key parameters:
- **`eps` (epsilon)**: The radius around a point to look for neighboring points
- **`min_samples`**: The minimum number of points within `eps` to form a dense region

There are three types of points:
- **Core Point**: Has at least `min_samples` points (including itself) within its `eps`-neighborhood
- **Border Point**: Has fewer than `min_samples` within `eps`, but is in the neighborhood of a core point
- **Noise Point**: Neither a core nor border point

DBSCAN clusters are formed by expanding dense regions from core points, connecting points within reach, and labeling outliers as noise.

#### Why Use DBSCAN?

- No need to guess k (number of clusters)
- Handles outliers naturally
- Ideal for non-spherical or uneven cluster shapes
- Works well when clusters have different densities

In this notebook, we apply DBSCAN to explore mental health profiles among university students. The goal is to group students based on similarities in their self-reported levels of academic stress, anxiety, and depression, without assuming a fixed number of clusters.

We use a standardized mental health dataset collected from 2,028 university students, processed from a Google Form survey with 26 Likert-scale psychological items.

### Required Libraries

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

### Load and Preprocess Data
We begin by loading the **Mental Health Assessment of University Students** dataset.

- It includes responses from **2,028 students** across various universities.
- The survey used **5-point Likert scales** to measure academic stress, anxiety, and depression symptoms.
- Each question was phrased as:  
  *"In a semester, how often have you felt [symptom]?"*
- Example symptoms include:
  - Trouble sleeping
  - Feeling hopeless
  - Suicidal thoughts
  - Confidence in handling academic problems

For preprocessing, we will:
1. **Select** the 26 psychological survey items
2. **Map** Likert-scale responses to numeric values (0–4)
3. **Drop** any rows with responses like “Prefer not to say”
4. **Standardize** the data using `StandardScaler` to ensure all features contribute equally to clustering

This preprocessing ensures the data is clean, numeric, and ready for DBSCAN to analyze.

```python
df = pd.read_csv("MentalHealth.csv")

likert_columns = [
    '1. In a semester, how often have you felt upset due to something that happened in your academic affairs? ',
    '2. In a semester, how often you felt as if you were unable to control important things in your academic affairs?',
    '3. In a semester, how often you felt nervous and stressed because of academic pressure? ',
    '4. In a semester, how often you felt as if you could not cope with all the mandatory academic activities? (e.g, assignments, quiz, exams) ',
    '5. In a semester, how often you felt confident about your ability to handle your academic / university problems?',
    '6. In a semester, how often you felt as if things in your academic life is going on your way? ',
    '7. In a semester, how often are you able to control irritations in your academic / university affairs? ',
    '8. In a semester, how often you felt as if your academic performance was on top?',
    '9. In a semester, how often you got angered due to bad performance or low grades that is beyond your control? ',
    '10. In a semester, how often you felt as if academic difficulties are piling up so high that you could not overcome them? ',
    '1. In a semester, how often you felt nervous, anxious or on edge due to academic pressure? ',
    '2. In a semester, how often have you been unable to stop worrying about your academic affairs? ',
    '3. In a semester, how often have you had trouble relaxing due to academic pressure? ',
    '4. In a semester, how often have you been easily annoyed or irritated because of academic pressure?',
    '5. In a semester, how often have you worried too much about academic affairs? ',
    '6. In a semester, how often have you been so restless due to academic pressure that it is hard to sit still?',
    '7. In a semester, how often have you felt afraid, as if something awful might happen?',
    '1. In a semester, how often have you had little interest or pleasure in doing things?',
    '2. In a semester, how often have you been feeling down, depressed or hopeless?',
    '3. In a semester, how often have you had trouble falling or staying asleep, or sleeping too much? ',
    '4. In a semester, how often have you been feeling tired or having little energy? ',
    '5. In a semester, how often have you had poor appetite or overeating? ',
    '6. In a semester, how often have you been feeling bad about yourself - or that you are a failure or have let yourself or your family down? ',
    '7. In a semester, how often have you been having trouble concentrating on things, such as reading the books or watching television? ',
    "8. In a semester, how often have you moved or spoke too slowly for other people to notice? Or you've been moving a lot more than usual because you've been restless? ",
    '9. In a semester, how often have you had thoughts that you would be better off dead, or of hurting yourself? '
]

likert_mapping = {
    "0 - Not at all": 0,
    "0 - Never": 0,
    "1 - Several days": 1,
    "1 - Several days (less than 15 days)": 1,
    "1 - Almost Never": 1,
    "2 - Sometimes": 2,
    "2 - More than half the days": 2,
    "2 - More than half the semester": 2,
    "3 - Fairly Often": 3,
    "3 - Nearly every day": 3,
    "4 - Very Often": 4,
    "4 - Prefer not to say": np.nan
}

df_numeric = df[likert_columns].replace(likert_mapping).infer_objects()
df_numeric_clean = df_numeric.dropna()
```

### Running DBSCAN and Visualizing Clusters with PCA

Once the data is cleaned and standardized, we apply the **DBSCAN clustering algorithm**. Here's the step-by-step process:

1. Standardize the Data
We use `StandardScaler` to scale each feature to have a mean of 0 and standard deviation of 1. This is crucial because DBSCAN uses **Euclidean distance**, and unscaled features would bias the results.

2. Apply DBSCAN
We fit DBSCAN using:
- `eps = 1.2`: The radius around each point to search for neighbors
- `min_samples = 5`: The minimum number of points required to form a dense region (a cluster core)

DBSCAN then:
- Groups together core points and their density-reachable neighbors
- Labels remaining points as **noise** with the value `-1`

3. Reduce Dimensions with PCA
Since our dataset has 26 features, we use **Principal Component Analysis (PCA)** to project the data down to 2 dimensions for easy visualization.

4. Visualize the Results
We use `seaborn` to create a scatter plot of the PCA-reduced data:
- Each point represents a student
- Color indicates the DBSCAN-assigned cluster
- Cluster `-1` denotes noise points (outliers)

This plot gives an intuitive view of how DBSCAN has segmented the student population based on psychological symptom similarity.


```python
# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numeric_clean)

# DBSCAN clustering
dbscan = DBSCAN(eps=1.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Reduce to 2D with PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Prepare DataFrame for plotting
dbscan_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
dbscan_df["Cluster"] = dbscan_labels

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dbscan_df, x="PC1", y="PC2", hue="Cluster", palette="tab10", s=60, alpha=0.8)
plt.title("DBSCAN Clustering Visualized with PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Cluster", loc="best")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Evaluating DBSCAN with Silhouette Score

To evaluate how well DBSCAN has clustered the data, we use the **Silhouette Score**, which measures how similar each point is to its own cluster versus other clusters. The score ranges from -1 (poor) to 1 (ideal), with values above 0.5 considered strong clustering.

Method:
- We exclude noise points (label = -1) from the calculation
- The silhouette score is computed only on the remaining clustered points
- This gives us a measure of cohesion vs separation among valid clusters

Result:
- In our case, the silhouette score is low due to:
  - Very few clusters being formed
  - Many points being marked as noise
  - Dense overlapping in PCA space


```python
# Mask out noise points (label = -1)
non_noise_mask = dbscan_labels != -1

# Only calculate silhouette score if there are at least 2 non-noise clusters
if non_noise_mask.sum() > 1 and len(set(dbscan_labels[non_noise_mask])) > 1:
    score = silhouette_score(X_scaled[non_noise_mask], dbscan_labels[non_noise_mask])
    print(f"Silhouette Score (excluding noise): {score:.3f}")
else:
    print("Too few non-noise clusters to compute silhouette score.")
```

### Conclusion

DBSCAN provided a unique perspective on the student mental health dataset. Unlike K-Means, which grouped all students into broad clusters, DBSCAN revealed that **the vast majority of students do not fall into tightly packed psychological subgroups**.

**Technical Interpretation:**
- **1,944 students** were labeled as **noise**, indicating they did not belong to any dense region in the feature space.
- Only **84 students** were clustered into small, high-density pockets.
- Despite the low number of clusters, the **silhouette score of 0.858** shows that these few clusters were **highly cohesive and well-separated** — indicating strong internal similarity within those groups.

**Social Implications:**
- The fact that most students were considered **noise** suggests that **mental health symptom patterns vary continuously** across the population rather than falling into clear categories.
- This supports a **dimensional model of mental health**, where students are distributed along spectrums of stress, anxiety, and depression, rather than fitting into discrete profiles.
- However, DBSCAN did identify a small subset of students whose symptoms are so similar and severe that they form **tight, rare clusters** — likely corresponding to students at **high risk** or experiencing **specialized psychological conditions**.

These insights highlight DBSCAN’s power as an outlier detection tool and reinforce the importance of **individualized mental health support**, especially for students who may not fit broad categories but still need care.


## $k$-Means vs DBSCAN

We applied both **$k$-Means** and **DBSCAN** to the same dataset of student mental health survey responses. Each method revealed different insights based on their underlying clustering logic.

| Criterion                  | $k$-Means                                  | DBSCAN                                                  |
|---------------------------|-------------------------------------------|----------------------------------------------------------|
| **Clustering Basis**      | Distance to centroids                     | Density of points in local neighborhoods                 |
| **Number of Clusters**    | Must be specified in advance (`k=3`)      | Determined automatically based on `eps` and `min_samples` |
| **Outlier Detection**     | No                                        | Yes — noise points labeled as `-1`                      |
| **Cluster Shape**         | Spherical / convex                        | Arbitrary (non-linear, non-convex shapes possible)       |
| **Result Overview**       | All students assigned to 1 of 3 large clusters | 7 small clusters + 1,944 students labeled as noise      |
| **Interpretability**      | Easy to profile and compare               | Best for rare pattern detection and outlier segmentation |
| **Silhouette Score**      | Not calculated (but clusters well-spaced) | **0.858 (excluding noise)** — excellent separation       |

### Key Takeaways:

- **$k$-Means** is better suited when:
  - You expect **broad patterns** and want clearly separated groups
  - Every point must be assigned a cluster
  - The data fits reasonably well to spherical distributions

- **DBSCAN** is better suited when:
  - You want to **detect outliers** or **rare patterns**
  - Clusters may have **uneven densities** or arbitrary shapes
  - You’re willing to leave many points unclustered (as noise)

In our case, **$k$-Means identified clear and interpretable profiles**, while **DBSCAN excelled at isolating extremely distinct patterns**, even if they were rare.

This contrast illustrates how combining methods can yield deeper insights into complex social data like student mental health.
