# Principal Component Analysis (PCA)

This notebook applies **Principal Component Analysis (PCA)** to the student mental health dataset to reduce dimensionality and visualize patterns in self-reported psychological symptoms.

PCA is an unsupervised linear transformation that projects high-dimensional data into a lower-dimensional space while preserving as much variance as possible.

We use PCA here to:
- Understand the underlying structure of the data
- Visualize students in 2D space based on symptom similarity
- Explore which combinations of symptoms explain the most variation in responses

We use the same mental health dataset used in previous clustering experiments. It contains responses from 2,028 students to 26 Likert-scale questions related to:

- Academic stress
- Anxiety symptoms
- Depression symptoms

Responses range from “0 - Not at all” to “4 - Very Often,” and are numerically mapped for analysis. We drop non-informative responses and standardize the features prior to applying PCA.


### Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading and Preprocessing the Data

We begin by loading the **Mental Health Assessment of University Students** dataset, which contains 2,028 student responses to 26 self-reported Likert-scale questions. These questions cover symptoms related to:

- **Academic stress** (e.g., "How often have you felt overwhelmed by academic responsibilities?")
- **Anxiety** (e.g., "How often have you felt nervous or on edge?")
- **Depression** (e.g., "How often have you felt down or had thoughts of self-harm?")

### Preprocessing Steps:

1. **Select Likert-scale features**  
   We isolate the 26 psychological items and exclude demographic columns (e.g., age, university).

2. **Map Likert responses to numeric values**  
   We convert responses like `"0 - Not at all"` and `"3 - Nearly every day"` into integers (0–4). This allows us to use mathematical transformations like PCA.

3. **Handle missing values**  
   Responses such as `"4 - Prefer not to say"` are mapped to `NaN` and removed to ensure clean numeric input.

4. **Standardize the data**  
   We apply `StandardScaler` to ensure all features have mean = 0 and standard deviation = 1. This is necessary because PCA is **sensitive to scale** and uses variance to compute principal components.

This results in a fully numeric, clean dataset ready for dimensionality reduction using PCA.

```python
df = pd.read_csv("MentalHealth.csv")

# Define Likert-based columns and mapping
likert_columns = [
    '1. In a semester, how often have you felt upset due to something that happened in your academic affairs? ',
    '2. In a semester, how often you felt as if you were unable to control important things in your academic affairs?',
    '3. In a semester, how often you felt nervous and stressed because of academic pressure? ',
    '4. In a semester, how often you felt as if you could not cope with all the mandatory academic activities? (e.g, assignments, quiz, exams) ',
    '5. In a semester, how often you felt confident about your ability to handle your academic / university problems?',
    '6. In a semester, how often you felt as if things in your academic life is going on your way? ',
    '7. In a semester, how often are you able to control irritations in your academic / university affairs? ',
    '8. In a semester, how often you felt as if your academic performance was on top?',
    '9. In a semester, how often you got angered due to bad performance or low grades that is beyond your control? ',
    '10. In a semester, how often you felt as if academic difficulties are piling up so high that you could not overcome them? ',
    '1. In a semester, how often you felt nervous, anxious or on edge due to academic pressure? ',
    '2. In a semester, how often have you been unable to stop worrying about your academic affairs? ',
    '3. In a semester, how often have you had trouble relaxing due to academic pressure? ',
    '4. In a semester, how often have you been easily annoyed or irritated because of academic pressure?',
    '5. In a semester, how often have you worried too much about academic affairs? ',
    '6. In a semester, how often have you been so restless due to academic pressure that it is hard to sit still?',
    '7. In a semester, how often have you felt afraid, as if something awful might happen?',
    '1. In a semester, how often have you had little interest or pleasure in doing things?',
    '2. In a semester, how often have you been feeling down, depressed or hopeless?',
    '3. In a semester, how often have you had trouble falling or staying asleep, or sleeping too much? ',
    '4. In a semester, how often have you been feeling tired or having little energy? ',
    '5. In a semester, how often have you had poor appetite or overeating? ',
    '6. In a semester, how often have you been feeling bad about yourself - or that you are a failure or have let yourself or your family down? ',
    '7. In a semester, how often have you been having trouble concentrating on things, such as reading the books or watching television? ',
    "8. In a semester, how often have you moved or spoke too slowly for other people to notice? Or you've been moving a lot more than usual because you've been restless? ",
    '9. In a semester, how often have you had thoughts that you would be better off dead, or of hurting yourself? '
]

likert_mapping = {
    "0 - Not at all": 0, "0 - Never": 0,
    "1 - Several days": 1, "1 - Several days (less than 15 days)": 1, "1 - Almost Never": 1,
    "2 - Sometimes": 2, "2 - More than half the days": 2, "2 - More than half the semester": 2,
    "3 - Fairly Often": 3, "3 - Nearly every day": 3,
    "4 - Very Often": 4, "4 - Prefer not to say": np.nan
}

df_numeric = df[likert_columns].replace(likert_mapping).infer_objects()
df_numeric_clean = df_numeric.dropna()

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numeric_clean)
```

### Applying PCA

We reduce the 26-dimensional dataset to:
- **2 components** for visualization
- **Full component set** to analyze explained variance

Each principal component is a linear combination of the original features that captures orthogonal directions of maximum variance in the data.

```python
pca = PCA()
X_pca_full = pca.fit_transform(X_scaled)

# Scree plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
         np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title("Cumulative Explained Variance by Principal Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance Explained")
plt.grid(True)
plt.tight_layout()
plt.show()
```

The scree plot above shows the **cumulative proportion of total variance** explained as we include more principal components.

**Interpretation:**

- The first few components capture a significant amount of information:
  - The first component alone explains over 40% of the total variance
  - The first 5 components explain over 60%
  - It takes about 10 components to explain 80–85% of the variance

**Why This Matters:**

- PCA is most valuable when a **small number of components** explain most of the variance
- In this case, **dimensionality can be reduced** substantially (e.g., to 2–10 components) while still retaining most of the underlying structure of the data
- This enables simpler visualization, clustering, and interpretation without much loss of information

We will now project the dataset onto the **first two principal components** to visualize how students are distributed across this reduced space.

### PCA Results and Interpretation

- The **first few principal components** explain a large portion of variance in the dataset
- PCA helps us reduce complexity while preserving key information
- We can now use PCA for visualizing the student population and exploring structure in the data

```python
# Use only the first 2 components for plotting
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca_2d, columns=["PC1", "PC2"])

plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", alpha=0.7)
plt.title("Students Projected onto First 2 Principal Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Conclusion

Principal Component Analysis (PCA) provided valuable insight into the structure of student mental health data.

By reducing the original 26 Likert-scale features to just a few orthogonal components, PCA allowed us to:
- Identify which directions in the data explain the most variance
- Simplify complex psychological patterns into a more interpretable form
- Visualize the distribution of students in 2D space while retaining the core structure of their responses

### Key Takeaways:

- The **first two components** alone explained a significant portion of the total variance, making them useful for visualization and further analysis.
- The 2D projection showed a **dense central region**, suggesting that most students fall along a continuum of mental health symptoms.
- No clear, discrete clusters emerged in the PCA plot, reinforcing the idea that **student mental health exists on a spectrum** rather than in rigid categories.
- Outlier points at the edges of the PCA plot may represent students with **extreme or unusual combinations of symptoms**, and may warrant further attention.

### Broader Implications:

- PCA confirms the **dimensional nature** of psychological distress in university populations — students differ in degree, not in kind.
- This supports the use of **personalized interventions** rather than one-size-fits-all treatment categories.
- PCA also prepares the data for **advanced unsupervised learning techniques**, such as clustering or anomaly detection, by reducing noise and redundancy.

In sum, PCA not only reduced dimensionality but also revealed important social and psychological structure in the dataset — making it a valuable first step in understanding and supporting student mental health.
